```python
import pandas as pd
import sys
import re
from re import search

# Importing Files from Same Directory
from lyricVariables import lyric_list
from accentVariables import accent_dict

# Don't collapse Pandas Dataframes:
pd.set_option('display.max_rows', None, 'display.max_columns', None)
```

```python
#######
# Goal:
# Scansion Interpreter
# Feet (Line Length) Interpreter
# Prosidy Highlighter
# Perceived Pitch Interpreter
```

```python
# Primary Variables:
stressed = '/' # The part of the word that is emphasized
unstressed = '^' # The part of the word that is NOT emphasized
```

```python
line_accent_list = []

###########################
###########################
for line in lyric_list:
    # pattern = re.compile(r'(?<!\w)(' + '|'.join(re.escape(key) for key in dictionary.keys()) + r')(?!\w)')
    pattern = re.compile('|'.join(re.escape(key) for key in accent_dict.keys()))
    result = pattern.sub(lambda x: accent_dict[x.group()], line)
    line_accent_list.append(result)
    
# print(line_accent_list)
# print('\n') 

###########################
###########################
clean_line_accent_list = []
for line in line_accent_list:
    clean_line = re.sub(' +', ' ', line, count=0, flags=re.I)
    clean_line_accent_list.append(clean_line)

# print(clean_line_accent_list)
# print('\n') 

###########################
###########################
clean_start_line_accent_list = []
for line in clean_line_accent_list:
    clean_line = re.sub('^ ', '', line, count=0, flags=re.I)
    clean_start_line_accent_list.append(clean_line)

# print(clean_start_line_accent_list)
# print('\n') 

###########################
###########################
clean_end_line_accent_list = []
for line in clean_start_line_accent_list:
    clean_line = re.sub(' $', '', line, count=0, flags=re.I)
    clean_end_line_accent_list.append(clean_line)

# print(clean_end_line_accent_list)
# print('\n') 
    
###########################
###########################
merged_accent_list = []
for line in clean_end_line_accent_list:
    merged_line = re.sub(' ', '', line, count=0, flags=re.I)
    merged_accent_list.append(merged_line)

# print(merged_accent_list)
```

```python
# print(lyric_list)
# print('\n')

###########################
###########################
clean_line_lyric_list = []
for line in lyric_list:
    clean_line = re.sub(' +', ' ', line, count=0, flags=re.I)
    clean_line_lyric_list.append(clean_line)

# print(clean_line_lyric_list)
# print('\n')

###########################|
########################### 
clean_start_line_lyric_list = []
for line in clean_line_lyric_list:
    clean_line = re.sub('^ ', '', line, count=0, flags=re.I)
    clean_start_line_lyric_list.append(clean_line)

# print(clean_start_line_lyric_list)
# print('\n') 

###########################
###########################
clean_end_line_lyric_list = []
for line in clean_start_line_lyric_list:
    clean_line = re.sub(' $', '', line, count=0, flags=re.I)
    clean_end_line_lyric_list.append(clean_line)

# print(clean_end_line_lyric_list)
```

```python
lyrics_and_accents_df = pd.DataFrame(clean_end_line_lyric_list, columns =['Line Lyric'])
lyrics_and_accents_df['Line Lyric Accent'] = pd.DataFrame(merged_accent_list)
lyrics_and_accents_df = lyrics_and_accents_df.drop_duplicates()
print('stressed = /')
print('unstressed = ^')
lyrics_and_accents_df.sort_values(by='Line Lyric Accent', ascending=True)
```
